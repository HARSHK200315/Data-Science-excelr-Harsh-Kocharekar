```python
import numpy as np
from scipy import stats

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
data
```

```python
n = len(data)
mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence = 0.99
alpha = 1 - confidence

t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_crit * (sample_std / np.sqrt(n))

ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

(mean, sample_std, (ci_lower, ci_upper))
```

```python
pop_std = 0.2 
z_crit = stats.norm.ppf(1 - alpha/2)

margin_of_error_z = z_crit * (pop_std / np.sqrt(n))

ci_lower_z = mean - margin_of_error_z
ci_upper_z = mean + margin_of_error_z

(pop_std, (ci_lower_z, ci_upper_z))
```
